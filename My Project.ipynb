```python
import tensorflow as tf
```

```python
import keras
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,
                                shear_range=0.2,
                                 rotation_range=180,
                                 zoom_range=0.2)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
x_train=train_datagen.flow_from_directory(r'C:\pooja\Dataset\trainset',target_size=(64,64),batch_size=32,class_mode='binary')
```

```python
x_test=train_datagen.flow_from_directory(r'C:\pooja\Dataset\testset',target_size=(64,64),batch_size=32,class_mode='binary')
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.layers import Conv2D
```

```python
from tensorflow.keras.layers import MaxPool2D
```

```python
from tensorflow.keras.layers import Flatten
```

```python
model=Sequential()
```

```python
model.add(Conv2D( 32,3,3,input_shape=(64,64,3),activation='relu'))
```

```python
model.add(MaxPool2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units=128,activation='relu',kernel_initializer='random_uniform'))
```

```python
model.add(Dense(units=1,activation='sigmoid',kernel_initializer='random_uniform'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit_generator(x_train,steps_per_epoch=8,validation_data=x_test,epochs=128,validation_steps=2)
```

```python
model.save('Missing_1.h5')
```
